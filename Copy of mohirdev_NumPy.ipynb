#### 29. [1, 2, 3, 4, 5] vektor berilgan, har bir qiymat orasiga ketma-ket 3 ta nol raqamini yozuvchi dastur tuzing.


`ðŸ’¡Ishora` array[::4]

````python
a = np.array([1, 2, 3, 4, 5])
s = np.array([])

for i in range((a.size - 1)):
  c = np.append(a[i:i+1], [0,0,0])
  s = np.append(s, c)
s = np.append(s, a[-1])

s
````
